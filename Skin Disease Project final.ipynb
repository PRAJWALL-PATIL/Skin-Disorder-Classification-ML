## KNN

* K-Nearest Neighbors (KNN) is a type of supervised learning algorithm which is used for both regression and classification purposes, but mostly it is used for the latter. Given the dataset, with different classes, KNN tries to predict the correct class of test data by calculating the distance between the test data and all the training points. It then selects the k points which are closest to the test data. The closeness beween data points is typically measured using distance metrics such as Euclidean distance. Once the points are selected, the algorithm calculates the probability (in case of classification) of the test point belonging to the classes of the k training points and the regression problem, the predicted value is the mean of the k selected training points. In medical diagnosis problems, including skin disease classification, KNN is particularly useful because it relies on similarity patterns among patient records. When combined with proper feature scaling and balanced data, KNN can effectively capture complex relationships between clinical histopathological features, making it a valuable baseline model for comparison. 

```python
# Importing libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
# taking optimal k to determine how many nearest neighbors to create
# creating a list to store the error values for each k
error_rate=[]

for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
error_rate
```

```python
# Plotting the k-value and error rate

plt.figure(figsize=(10,8))
plt.plot(range(1,11), error_rate, color='green', linestyle = 'dashed',
        marker = 'o', markerfacecolor = 'red', markersize=10)
plt.title('Error Rate Vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse_val = [] #creating list to store rmse values for different k 
for K in range(1,25):
    model = KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train) #to fit the model
    pred = model.predict(X_test) #to make prediction on test set
    error = np.sqrt(mean_squared_error(y_test, pred)) #calculate rmse
    rmse_val.append(error) #store the rmse values
    print ('RMSE Value for k=', 'is:', error)
```

# SVM

* SVM is the Supervised Machine Learning algorithm used for both classification, regression. Given the dataset, the algorithm tries to divide the data using hyperplanes and then makes the predictions. SVM is a non-probabilistic linear classifier. SVM constructs a best line or the decision boundary called Hyperplane which can be used for classification or regression or outlier detection. This hyperplane creates 2 margin lines parallel to it which have some distance so that it can distinctly classify the data points. The distance between the 2 margin lines are called marginal distance. These 2 margin lines passes through the most nearest positive points and the most nearest negative points. Those points through which the margin lines pass are called support vectors. Support vectors are important as it helps to determine the maximum distance of the marginal plane.

# Decision Tree

* A decision tree is a supervised machine learning algorithm used for classification and regression tasks. It works by recursively splitting the dataset into subsets based on feature values, forming a tree-like structure of decision rules. Each internal mode represents a test on a feature, each branch corresponds to an outcome of the test, and each leaf node represents a final prediction. Decision trees are widely used due to their simplicity,ability to handle both numerical and categorical data and use of interpretation.

By introducing randomness in both data sampling and feature selection, random forest reduces overfitting and improves generalization compared to a single decision tree. # Suggestions to Doctors for Early Diagnosis of Skin Diseases

## Important Features Identified by the Model
* The machine learning models identified several clinical and histopathological features as key contributors to accurate skin disease classification. Features such as erythema, scaling, definite borders, itching, knee and elbow involvment, scalp involvment, and family history were consistently important. Histopathological indicators including acanthosis, parakeratosis, inflammatory inifltrate, and elongation of rete ridges also played a significant role in differentiating similar skin conditions

## Clinical and Histopathological Relevance
* Many erythemato-squamous diseases share common symptoms, making early diagnosis challenging. The integration of histopathological features helps distinguish conditions with overlapping clinical presentations. For instance, psoriasis shows characteristic scalp and knee involvement, while lichen planus is associated with polygonal papules and oral mucosal involvement. Recognising these subtle patterns can reduce the need for invasive biopsies in early stages.

## Suggestions for Early Identification by Doctors
* Priority should be given to patients presenting with persistent erythema and scaling that do not respond to initial treatment.
* To carefully evaluate lesion distribution, especially involvement of scalp, knees and elbows, which strongly indicate psoriosis.
* To record family history systematically, as genetic predisposition improves diagnostic confidence.
* To use histopathological features in combination rather than isolation to distinguish similar disease.
* Refer high risk or ambiguous cases for early machine learning assisted screening to avoid delayed daignosis.

## Role of Machine Learning in early detection
* Machine learning models assist doctors by analysing complex patterns across multiple features simultaneously. These models reduce subjectively, improve diagnostic consistency, and support decision-making, particularly in cases with overlapping clinical characteristics. Early prediction using machine learning enables timely intervention, reduces diseases progression, and improves patient outcomes.
